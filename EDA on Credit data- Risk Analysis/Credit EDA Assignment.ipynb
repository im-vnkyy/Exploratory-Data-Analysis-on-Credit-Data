```python
def outliers_treatment(column):
    Q1 = df_application[column].describe().loc["25%"]
    Q3 = df_application[column].describe().loc["75%"]
    IQR = Q3- Q1
    upper = Q3 + 1.5 *IQR
    lower = Q1 - 1.5 *IQR
    df_application[column] = np.where(df_application[column] > upper, upper, df_application[column])
    df_application[column] = np.where(df_application[column] < lower, lower, df_application[column])
```

```python
# outlier treatment on numerical variables which need to be capped and floored:
for i in ["AMT_INCOME_TOTAL", "AMT_CREDIT", "AMT_ANNUITY","AMT_GOODS_PRICE"]:
    outliers_treatment(i)
```
